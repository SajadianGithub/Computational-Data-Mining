```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pyperclip
import time

def get_distance(from_city, to_city):
    url = "https://www.distancefromto.net/"
    
    driver = webdriver.Chrome()  # Initialize your webdriver (make sure to have chromedriver installed)
    driver.get(url)

    try:
        from_input = driver.find_element(By.ID, "distancefrom")
        to_input = driver.find_element(By.ID, "distanceto")
        submit_button = driver.find_element(By.ID, "hae")

        from_input.send_keys(from_city)
        to_input.send_keys(to_city)
        submit_button.click()

        # time.sleep(3)

        distance_element = driver.find_element(By.ID, "totaldistancekm")

        distance_element.send_keys(Keys.CONTROL + 'a') 
        distance_element.send_keys(Keys.CONTROL + 'c')
        clipboard_content = pyperclip.paste()

        return int(float(clipboard_content[:-3]))
    
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        driver.quit()
```

```python
from collections import defaultdict 
data = defaultdict(dict) 

for i, from_city in enumerate(columns_):
    for j in range(i+1, len(columns_)):

        if (from_city == columns_[j]):
            data[from_city][columns_[j]] = 0
        else:
            distance = get_distance(from_city, columns_[j])
            print(distance)
            data[from_city][columns_[j]] = distance
```

```python
df = pd.DataFrame(data)
```
